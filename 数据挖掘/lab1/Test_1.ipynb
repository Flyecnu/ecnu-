```python
path="Data/data3/"
f_w_1=open("Data/data3/train.txt","a+",encoding="utf-8")
f_w_2=open("Data/data3/dev.txt","a+",encoding="utf-8")
f_w_3=open("Data/data3/test.txt","a+",encoding="utf-8")

for i in range(8):
    f_r=open(path+str(i)+".txt","r")
    
    m=f_r.readlines()
    x=int(len(m)*0.7)
    y=int(len(m)*0.85)
    print(x,y)
    print(len(m))
    
    for j in range(len(m)):
        temp=m[j].strip('\n')
        if j<x:
            f_w_1.write(temp+'\t'+str(i)+'\n')
        if (j>=x and j<y):
            f_w_2.write(temp+'\t'+str(i)+'\n')
        elif j>=y:
            f_w_3.write(temp+'\t'+str(i)+'\n')
    
    f_r.close()
    
f_w_1.close()
f_w_2.close()
f_w_3.close()
    
    
    
```

## 循环爬

```python
url="http://www.chinanews.com.cn"
html=requests.get(url)
soup=BeautifulSoup(html.content,"html.parser")
```

```python
url_scroll = soup.find(name='ul',attrs={"class":"nav_navcon"}).find('a')['href'] #获取顶部导航栏排
url_scroll = url + url_scroll # https://www.chinanews.com/scroll-news/news1.html
html_scroll = requests.get(url_scroll) #将string类型的网址转换成可爬取的url
soup_scroll = BeautifulSoup(html_scroll.content,"html.parser")
all_scroll_news = []
# href_id=[2,4,5,7,8,9,10,12,13,14]#时政、国际、社会、港澳、台湾、华人、财经、金融、房产、汽车、文化、娱乐、体育
category=["时政","国际","社会","港澳","台湾","华人","财经","文化","娱乐","体育"] #初始的13种新闻类别
urlid_scrool=soup_scroll.find(name='div',attrs={"id":"newsdh"}) #找到滚动新闻栏目
for c in category:
    print(c)
    tmplist=[]
    for e in urlid_scrool.find_all('a'):  #找到对应的类别的url
        if (e.string==c):
            catetory_url=url+e['href'] #类别网站的网址
            catetory_html=requests.get(catetory_url)
            catetory_scroll = BeautifulSoup(catetory_html.content,"html.parser",from_encoding="utf-8") 
            news = catetory_scroll.find(name="div", attrs={"class" :"content_list"}) #找到装有该类别新闻的content_list
            news = news.findAll(name="div",attrs={"class":"dd_bt"}) #获取content_list下所有的dd_bt：里面存有新闻全文的链接和新闻标题(string)
            for n in news:
#                 print(n.a.string.encode("iso-8859-1").decode("utf-8")) #编码转换
                tmplist.append(n.a.string.encode("utf-8").decode("utf-8"))
            all_scroll_news.append(tmplist) #按类别存放爬取的所有新闻标题
            print("len news:",len(all_scroll_news))
# all_scroll_news
```

## 国内新闻

```python
f=open("Data/data_new/0.txt","a+",encoding="UTF-8")


for i in range(300,1000):
    url="http://www.ahwang.cn/china/index_"+str(i)+".html"
    html=requests.get(url)
    soup=BeautifulSoup(html.content,"html.parser")

    all_news=[]

    important_news=soup.find_all(name="div",attrs={"class":"news_list clearfix"})

    for m in important_news:
        # important_news=m.find_all(name={"li"})+m.find_all(name={"a"})+m.find_all(name={"ul"})
        n=m.find_all(name={"a"})
        for x in n:
            if (x.string):
                all_news.append(x.string)
    
    for n in all_news:
        f.write(n+'\n')
f.close()
```
