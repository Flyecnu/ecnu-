```python
# 最终版
import tkinter as tk
from tkinter import filedialog, messagebox
import subprocess

class CompilerApp:
    def __init__(self, root):
        self.root = root
        self.root.title("CX Compiler")
        self.root.geometry("1200x900")
        self.root.minsize(800, 600)

        # 创建菜单栏
        menu_bar = tk.Menu(root)
        file_menu = tk.Menu(menu_bar, tearoff=0)
        file_menu.add_command(label="导入", command=self.import_file)
        file_menu.add_command(label="保存", command=self.save_file)
        menu_bar.add_cascade(label="文件", menu=file_menu)
        root.config(menu=menu_bar)
        # 创建顶部容器
        top_frame = tk.Frame(root)
        top_frame.pack(fill=tk.BOTH, padx=10, pady=10, expand=True)
        x = int(0.8 * root.winfo_screenwidth())

        # 创建左侧容器
        left_frame = tk.Frame(top_frame,width=100)
        left_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

        # 创建代码框
        code_label = tk.Label(left_frame, text="代码框")
        code_label.pack()
        self.code_text = tk.Text(left_frame, width=100, height=20, bg="white", bd=1, relief=tk.SOLID)
        self.code_text.pack(padx=5, pady=5, expand=True)

        # 创建右侧容器
        right_frame = tk.Frame(top_frame,width=100)
        right_frame.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)

        # 创建读入文本框
        input_label = tk.Label(right_frame, text="读入")
        input_label.pack()
        self.input_text = tk.Text(right_frame, width=100, height=20, bg="white", bd=1, relief=tk.SOLID)
        self.input_text.pack(padx=5, pady=5, expand=True)


        # 创建底部容器
        bottom_frame = tk.Frame(root)
        bottom_frame.pack(fill=tk.BOTH, padx=10, pady=10, expand=True)

        # 创建左侧容器
        left_frame = tk.Frame(bottom_frame)
        left_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

        # 创建目标指令显示框
        instructions_label = tk.Label(left_frame, text="目标指令显示框")
        instructions_label.pack()

        self.instructions_text = tk.Text(left_frame, width=40, height=15, bg="white", bd=1, relief=tk.SOLID)
        self.instructions_text.pack(fill=tk.BOTH, padx=5, pady=5, expand=True)

        # 创建数据管理显示框
        data_label = tk.Label(left_frame, text="数据管理显示框")
        data_label.pack()

        self.data_text = tk.Text(left_frame, width=40, height=15, bg="white", bd=1, relief=tk.SOLID)
        self.data_text.pack(fill=tk.BOTH, padx=5, pady=5, expand=True)

        # 创建右侧容器
        right_frame = tk.Frame(bottom_frame)
        right_frame.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)

        # 创建错误提示框
        error_label = tk.Label(right_frame, text="错误提示框")
        error_label.pack()

        self.error_text = tk.Text(right_frame, width=40, height=5, bg="white", bd=1, relief=tk.SOLID)
        self.error_text.pack(fill=tk.BOTH, padx=5, pady=5, expand=True)

        # 创建运行结果显示框
        result_label = tk.Label(right_frame, text="运行结果显示框")
        result_label.pack()

        self.result_text = tk.Text(right_frame, width=40, height=5, bg="white", bd=1, relief=tk.SOLID)
        self.result_text.pack(fill=tk.BOTH, padx=5, pady=5, expand=True)

        # 创建按钮容器
        button_frame = tk.Frame(root)
        button_frame.pack(side=tk.BOTTOM, padx=10, pady=10)

        # 创建导入按钮
        import_button = tk.Button(button_frame, text="导入", command=self.import_file, bg="blue", width=10, height=2)
        import_button.pack(side=tk.LEFT, padx=5)

        # 创建导出按钮
        export_button = tk.Button(button_frame, text="导出", command=self.save_file, bg="blue", width=10, height=2)
        export_button.pack(side=tk.LEFT, padx=5)

        # 创建编译运行按钮
        compile_button = tk.Button(button_frame, text="编译运行", command=self.compile_and_run, bg="blue", width=10, height=2)
        compile_button.pack(side=tk.LEFT, padx=5)

        root.update()  # 更新窗口大小

    def import_file(self):
        filepath = filedialog.askopenfilename(filetypes=[("CX Files", "*.cx")])
        if filepath:
            with open(filepath, "r") as file:
                code = file.read()
                self.code_text.delete("1.0", tk.END)
                self.code_text.insert(tk.END, code)

    def save_file(self):
        code = self.code_text.get("1.0", tk.END)
        filepath = filedialog.asksaveasfilename(defaultextension=".cx", filetypes=[("CX Files", "*.cx")])
        if filepath:
            with open(filepath, "w") as file:
                file.write(code)
                messagebox.showinfo("保存成功", "文件保存成功！")

    def compile_and_run(self):
        code = self.code_text.get("1.0", tk.END)
        input_data = self.input_text.get("1.0", tk.END)

        # 清空文件内容
        files = ["fstack.txt", "fcode.txt", "foutput.txt", "fresult.txt", "ftable.txt"]
        for file_name in files:
            with open(file_name, "w") as file:
                file.write("")

        # 将代码保存到临时文件中
        with open("test.cx", "w") as file:
            file.write(code)

        # 将输入数据保存到临时文件中
        with open("input.txt", "w") as file:
            file.write(input_data)

        # 运行编译命令
        command = "./cx -c -t -s test.cx < input.txt"
        subprocess.run(command, shell=True)

        # 读取生成的文件内容
        files = ["fstack.txt", "fcode.txt", "foutput.txt", "fresult.txt", "ftable.txt"]
        for file_name in files:
            try:
                with open(file_name, "r") as file:
                    content = file.read()
                    if file_name == "fstack.txt":
                        self.data_text.delete("1.0", tk.END)
                        self.data_text.insert(tk.END, content)
                    elif file_name == "fcode.txt":
                        self.instructions_text.delete("1.0", tk.END)
                        self.instructions_text.insert(tk.END, content)
                    elif file_name == "foutput.txt":
                        self.error_text.delete("1.0", tk.END)
                        self.error_text.insert(tk.END, content)
                    elif file_name == "fresult.txt":
                        self.result_text.delete("1.0", tk.END)
                        self.result_text.insert(tk.END, content)
            except FileNotFoundError:
                self.error_text.delete("1.0", tk.END)
                self.error_text.insert(tk.END, f"未找到文件：{file_name}")

    def run(self):
        self.root.mainloop()

# 创建主窗口
root = tk.Tk()
app = CompilerApp(root)
app.run()
```
